## Resumen del Examen

La actividad principal consiste en trabajar con datos: extraerlos, procesarlos, transferirlos, transformarlos, almacenarlos y analizarlos.

Este examen propone desarrollar un pequeño proyecto de datos utilizando información de pares de divisas del mercado Forex, incluyendo registros diarios de precios:

- Open  
- High  
- Low  
- Close  
- Volumen  

### Flujo del proyecto

1. Extracción automática de datos mediante una API de mercado  
2. Análisis exploratorio para comprender la dinámica de precios  
3. Entrenamiento de un modelo de machine learning  
4. Generación de decisiones operativas diarias  
5. Encapsulamiento del pipeline completo en Docker  

Este escenario simula un entorno real donde pueden existir distintos tipos de datos, siendo las series temporales solo uno de ellos.
