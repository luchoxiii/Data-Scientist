# Caminata Aleatoria / Random Walk

<b> Import Librarys</b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
```

<b>Definicion </b>

Una caminata aleatoria o random walk, la podemos generar de manera muy sencilla, haciendo uso de la distribución normal estándar (media cero y varianza uno). Les dejo aquí el código en Python para los que quieran saber cómo se generaría, fíjense que al final hacemos un plot, y si esto lo corren varias veces (o dentro de un ciclo for) verán que siempre obtienen un camino diferente:

<b> Codigo</b>

```python
# Generamos 500 pasos aleatorios para la caminara, con media 0 y desv tipica 1
steps = np.random.normal(loc=0, scale=1, size=500)
```

```python
# Ponemos el primer elemento a 0 para que el precio inicial sea el que pongamos nosotros
steps[0]=0
```

```python
# Simulando los precios P con un precio inicial de 100
P = 100 + np.cumsum(steps)
```

```python
# Plot
plt.plot(P)
plt.title("Random Walk Simulada")
plt.show()
```
